```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
img = r'C:\Users\user\Desktop\30203\3.png'
```

```python
Image.open(img)
```

```python
text = open(r'C:\Users\user\Desktop\30203\we.txt','r',encoding = 'utf8').read()
```

```python
text
```

```python
lp_mask = np.array(Image.open(img))
lp_mask
```

```python
lp_wc=WordCloud(
    font_path = r'‪C:\Windows\Fonts\gulim.ttc',
background_color = 'white', 
    mask = lp_mask, 
    max_font_size=50, 
    contour_width=5, 
    contour_color='skyblue').generate(text)
```

```python
lp_wc.words_
```

```python
lp_wc.to_image()
```
